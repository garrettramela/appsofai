```python
# Imports the pandas library and sets it as alias pd.
import pandas as pd
```

```python
# Reads the data set 'sales.csv' into variable df (for data frame) using pandas and indexing the 1st column month.
df = pd.read_csv('sales.csv', index_col='month')
```

```python
# Viewing the data set 'sales.csv' by calling the variable df.
df
```

```python
# An example of indexing using square bracket notation.
df['salt']['Jan']
```

```python
# An example of using the loc (locator) method in pandas. Notice the field and row names switch from the previous example.
df.loc['May', 'spam']
```

```python
# Python indexes starting with 0 instead of 1 like in the R programming language. This example uses the iloc (purely integer-location based indexing) to locate the same 52 value as in the previous example.
df.iloc[4, 2]
```

```python
# Explore selecting only some columns by creating a new data frame and selecting the desired columns in bracket notation.
df_new = df[['salt', 'eggs']]
```

```python
# View the new data frame with the two columns salt and eggs.
df_new
```

```python
# This example shows how to select a singular column. Notice the column name and data type shows at the bottom.
df['eggs']
```

```python
# TBD
type(df['eggs'])
```

```python
# This example selects the singular column eggs and takes indexed rows 1:4 (Feb, Mar, Apr). Note: it doesn't include Jan because it is indexed at 0.
df['eggs'][1:4]
```

```python
# This example shows a selection on a singular column and a singular indexed row.
df['eggs'][4]
```

```python
# This example uses a colon (:) to select all rows and all columns between eggs and salt.
df.loc[:, 'eggs':'salt']
```

```python
# This example uses the integer-based locator (iloc) pandas method to select the indexed rows 2 through 
# 5 (Mar-May) and all indexed columns from 1 thereafter (Salt).
df.iloc[2:5, 1:]
```

```python
# This example uses the integer-based locator (iloc) pandas method to select rows 0 (Jan), 4 (May), and 5 (Jun) 
# and all indexed columns 0:2
df.iloc[[0, 4, 5], 0:2]
```

```python
# This example selects the salt column and uses booleans to show whether salt's values are 
# greater than 60 or less than 60.
df.salt > 60
```

```python
df[df.salt > 60]
```

```python
enough_salt_sold = (df.salt > 60)
```

```python
enough_salt_sold
```

```python
# This example shows how to combined filters using the ampersand operator which denotes AND.
df[(df.salt >= 50) & (df.eggs < 200)]
```

```python
# This example shows how to combined filters using the pipe operator which denotes OR.
df[(df.salt >= 50) | (df.eggs < 200)]
```

```python
user_usage = pd.read_csv('user_usage.csv')
```

```python
# 1st set of brackets are used to initialize a list then the 2nd set is used to switch the column names.
user_usage[['use_id', 'outgoing_mins_per_month', 'outgoing_sms_per_month', 'monthly_mb']]
```

```python
user_device = pd.read_csv('user_device.csv')
```

```python
user_device
```

```python
user_usage.head()
```

```python
user_device.head()
```

```python
result = pd.merge(user_usage,
                  user_device[['use_id', 'platform', 'device']],
                  on = 'use_id')
```

```python
result.head()
```

```python
# By default, pandas merge acts as an inner join wherein Python only keeps the unique values. 
# There are four merge types.
print('user usage dimensions: {}'.format(user_usage.shape))
print('user device dimensions: {}'.format(user_device[['use_id', 'platform', 'device']].shape))
print('user result dimensions: {}'.format(result.shape))
```
